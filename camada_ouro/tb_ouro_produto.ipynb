```python
import sys
sys.path.append('../lib') 
from gerenciamento_banco import *
from ExtracaoDados import *
from transformacao import *
import matplotlib.pyplot as plt
import seaborn as sns
from graficos import *
```

```python
extracao = ExtracaoDados()
df = extracao.carregar_tabela_sqlite3(tabela = "produto", nome_banco = "../banco_dados/db_prata.db")
df.head()
```

```python
categoria_count = df['categoria'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=categoria_count.index, y=categoria_count.values, palette='viridis')
plt.title('Número de Produtos por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Número de Produtos')
plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.pie(categoria_count, labels=categoria_count.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribuição de Produtos por Categoria')
plt.show()
```
